```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH_DEPLOYMENT = './simulation_setups/Deployment_within_550.npy'
dev_deployment = np.load(PATH_DEPLOYMENT)
```

```python
theta = np.linspace(0, 2*np.pi, 100)

figure, axes = plt.subplots(1, figsize=(10,10))
axes.set_aspect(1)

radius = 550
a = radius*np.cos(theta)
b = radius*np.sin(theta)
axes.plot(a, b, linestyle='dashed', linewidth=0.5, color='k')

radius = 400
a = radius*np.cos(theta)
b = radius*np.sin(theta)
axes.plot(a, b, linestyle='dashed', linewidth=1, color='k')

radius = 200
a = radius*np.cos(theta)
b = radius*np.sin(theta)
axes.plot(a, b, linestyle='dashed', linewidth=0.5, color='k')

axes.plot(0, 0, 'b^', label='Base Station',markersize=10)
axes.plot(dev_deployment[:,0], dev_deployment[:,1], 'r.', label='Device', color='r')

#plt.title('Circle using Parametric Equation')
plt.legend(loc='lower right')
plt.show()
```
